### Get Ranking Data

```python
import os
import sys
import json
import time
import datetime
from importlib import reload

import requests

import pandas as pd
import numpy as np

import external as ext
```

```python
with open('telemetry_data/2020-04-09/pc/4c76d82b-90f7-4fd1-9ceb-97b99347e94e.json', 'r') as file:
    json_data = file.read()
    file.close()
data = []
matchdef = []
for i in json.loads(json_data):
    if i["_T"] in ("LogMatchEnd"):
        data.append(i)
    if i["_T"] in ("LogMatchDefinition"):
        matchdef.append(i)
```

```python
matchdef[0]
```

```python
data
```

```python
team = []
rank  = []
for i in data[0]['characters']:
    team.append(i['character']['teamId'])
    rank.append(i['character']['ranking'])
    
ranking = pd.DataFrame()
ranking = pd.concat((pd.DataFrame(team, columns=['teamId']), pd.DataFrame(rank, columns = ['ranking'])), axis=1)
ranking = ranking.drop_duplicates().sort_values(by='teamId').reset_index().drop(['index'], axis=1)
indexes = []
for team in ranking.teamId.unique():
    indexes.append(ranking[ranking['teamId'] == team]['ranking'].idxmin())
ranking = ranking[ranking.index.isin(indexes)].reset_index().drop(['index'], axis=1)
ranking['matchId'] = matchdef[0]["MatchId"]
ranking = ranking.reindex(columns = ["matchId", "teamId", "ranking"])
ranking
```
