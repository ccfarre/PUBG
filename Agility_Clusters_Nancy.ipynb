```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('combined_csv.csv',nrows=800000)
df = df[df.distance1_2 != -999]
```

## Sorting Clusters

```python
matrix=df[['cluster_A', 'cluster_B','cluster_C','cluster_D']].values
matrix=np.sort(matrix,axis=1)
df.loc[:,['cluster_A', 'cluster_B','cluster_C','cluster_D']]=matrix
```

## Feature Engineering Agility

```python
df['agility']=np.where(df['strategy']!=df['strategy'].shift(-1), 1,0)
agility_init=df.groupby('teamId')['agility'].value_counts()
agility_perc=agility_init.div(agility_init.groupby('teamId').transform('sum')).to_frame().unstack() 
agility_perc=(agility_perc['agility'][1]*100).round(2).to_frame(name='agility')
df=pd.merge(agility_perc, df, on='teamId', how='left')
```
